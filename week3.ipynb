```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer,LabelEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer, KNNImputer

df = pd.read_csv('./Datasets/credit_scoring_eng.csv')
df
```

```python
imputer=SimpleImputer(strategy="mean")
cols=df.select_dtypes(include=["number"]).columns
df[cols]=imputer.fit_transform(df[cols])
df
```

```python
imputer=KNNImputer(n_neighbors=5)
df['days_employed']=imputer.fit_transform(df[['days_employed']])
df
```

```python
samp_df = df.copy()
samp_df['days_employed']=samp_df['days_employed'].fillna(method='ffill')
samp_df
```

```python
scaler=MinMaxScaler()
df[["days_employed","total_income"]]=scaler.fit_transform(df[['days_employed','total_income']])
df
```

```python
cols=df.select_dtypes(include=["number"]).columns
scaler=StandardScaler()
df[cols]=scaler.fit_transform(df[cols])
df
```

```python
norm=Normalizer(norm="l2")
cols=df.select_dtypes(include=["number"]).columns
df[cols]=norm.fit_transform(df[cols])
df
```

```python
encoder=LabelEncoder()
df['gender_encoded']=encoder.fit_transform(df['gender'])
df
```

```python
encoder=OrdinalEncoder(categories=[["F",'M','XNA'],["civil partnership",'divorced','married','unmarried','widow / widower']])
df[['gender_1','family_status_1']]=encoder.fit_transform(df[['gender','family_status']])
df
```

```python

print(pd.get_dummies(df['gender'],prefix="i"))
```

```python

print(pd.get_dummies(df, columns=["gender", "family_status"], prefix=["i", "m"]))
```

```python
df["days_employed_binned"]=pd.cut(df['days_employed'],bins=4,labels=["v low","low","medium","high"])
df
```

```python
custom_bins=[10,20,30,40,50]
custom_labels=["v low","low","medium","high"]
df["days_employed_cbinned"]=pd.cut(df["days_employed"],bins=custom_bins,labels=custom_labels)
df
```
